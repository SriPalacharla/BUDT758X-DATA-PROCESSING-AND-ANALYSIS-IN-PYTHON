c. Based on your answer from part b), revise your import process from part a) so that you can use the dictionary to create a more appropriately formatted DataFrame. Demonstrate your improved process by displaying (printing) your modified dictionary format and previewing the first 10 rows of your resultant DataFrame.

```python
path = 'F:/MS in Business Analytics/Spring/Python/Data Files/'
fh = open(path + 'inc5000_2017.csv')
reader = csv.reader(fh)
rank = 0
inc100 = {'rank':[], 'name':[], 'founded':[], 'industry':[], 'metro':[], 'workers':[]}
for line in reader :
    if rank == 0 :
        rank += 1
        continue
    if rank <= 100 :
        inc100['rank'].append(rank)
        inc100['name'].append(line[-1])
        inc100['founded'].append(line[-2])
        inc100['industry'].append(line[6])
        inc100['metro'].append(line[3])
        inc100['workers'].append(line[5])
        rank += 1   
print(inc100)
```

```python
df = pd.DataFrame(inc100)
df.set_index('rank', inplace=True)
df.head(10)
```

h.  Explore the types of industries in some of the top 20 cities. Specifically, write a function sum_industry that takes in a DataFrame (df) and a city (default to New York City), and returns the frequency counts of the industry column. Use the function to summarize the industry concentration for New York City, Atlanta, Chicago, and Washington (DC). What observations can you make about the industry concentration in each of these markets?

```python
def sum_industry(df, city='New York City') :
    return df[df['city'] == city]['industry'].value_counts()
```

```python
sum_industry(df=df, city='New York City')
```

```python
sum_industry(df=df, city='Atlanta')
```

```python
sum_industry(df=df, city='Chicago')
```

```python
sum_industry(df=df, city='Washington')
```

We can cleary observe that in New York, Atlanta and Chicago, the 3 most dominant industries are Advertising & Marketing, Business Products & Services and IT Services. But in Washington DC although Business Products & Services and Advertising & Marketing are quite dominant, the leading industry is Government Services.
