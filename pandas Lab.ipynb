### Exercise #2

Write a function top_sellers that takes in the DataFrame of video game sales (df), a category (cat, default=None), value (val, default=None), and a number (n, default=5) that specifies the number of rows to display in the result. The function should return a DataFrame containing the top n video games (in terms of global sales) for a specific value in the category (if specified) or the top n video games overall (if cat and val are not specified).

```python
def top_sellers(df=data, cat=None, val=None, n=5):
    if cat != None and val != None:
        return data[data[cat]==val].head(n) 
    else : 
        return data.head(n)
```

Use the function to display the following top selling games:

Top 5 overall sellers

```python
top_sellers(n=5)
```

Top 10 NES games

```python
top_sellers(cat='Platform',val='NES',n=10)
```

Top 10 selling games in the year 2000

```python
top_sellers(cat='Year',val=2000,n=10)
```

Top 10 selling sports games 

```python
top_sellers(cat='Genre',val='Sports',n=10)
```

### Exercise #3

Generate a (linear) correlation matrix between sales in specific markets (North America, Europe, Japan, Other) and global sales. 

```python
data[['NA_Sales','EU_Sales','JP_Sales','Other_Sales','Global_Sales']].corr()
```

What do you observe about the relationship between sales in these various markets?

North America Sales & Europe Sales highly correlate with a positive 0.76. There is also high correlation between Europe and Other Sales with 0.72. North America also correlates well with Other Sales with 0.63. Japan does have positive correlation with North America, Europe and Other but with very small values of 0.44, 0.43 and 0.29 respectively.

North America and Europe Sales correlate very highly with global sales followed by Other and Japan Sales but with comparatively lesser values.
