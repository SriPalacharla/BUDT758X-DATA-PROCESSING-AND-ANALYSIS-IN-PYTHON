```python
def play(games=1) :
    summary = {'Player 1 Initial' : [], 'Player 2 Initial' : [], 'Player 1 Final' : [], 
                'Player 2 Final' : [], 'Number of tie rounds' : [], 'Largest Round Total' : []}
    for z in range(games) :
        k = deal()
        P1 = k[0]
        P2 = k[1]

        init_total_P1 = sum(P1)
        init_total_P2 = sum(P2)

        X1= list(zip(P1[::2], P1[1::2]))
        X2= list(zip(P2[::2], P2[1::2]))
        P1_new = []
        P2_new = []
        tie = []

        tie_rounds = 0
        largest_round_total = 0


        for i in range(len(X1)):
            h1 = X1[i]
            h2 = X2[i]
            s1 = sum(h1)
            s2 = sum(h2)
            a1, a2 = h1
            b1, b2 = h2
            if s1 > s2 :
                P1_new.extend([a1,a2,b1,b2])
                if a1+a2+b1+b2 > largest_round_total :
                    largest_round_total = a1+a2+b1+b2
                if len(tie) > 0 :
                    P1_new.extend(tie)
                    tie = []
                    if sum(tie) > largest_round_total :
                        largest_round_total = sum(tie)
            elif s2 > s1 :
                P2_new.extend([a1,a2,b1,b2])
                if a1+a2+b1+b2 > largest_round_total :
                    largest_round_total = a1+a2+b1+b2
                if len(tie) > 0 :
                    P2_new.extend(tie)
                    tie = []
                    if sum(tie) > largest_round_total :
                        largest_round_total = sum(tie)
            elif s1 == s2 :
                tie.extend([a1,b1,a2,b2])
                if sum(tie) > largest_round_total :
                        largest_round_total = sum(tie)
                tie_rounds += 1
                if i == len(X1)-1 :
                    P1_new.extend(tie[::2])
                    P2_new.extend(tie[1::2])
                    if sum(tie) > largest_round_total :
                        largest_round_total = sum(tie)

        final_total_P1 = sum(P1_new)
        final_total_P2 = sum(P2_new)
        
        summary['Player 1 Initial'].append(init_total_P1)
        summary['Player 2 Initial'].append(init_total_P2)
        summary['Player 1 Final'].append(final_total_P1)
        summary['Player 2 Final'].append(final_total_P2)
        summary['Number of tie rounds'].append(tie_rounds)
        summary['Largest Round Total'].append( largest_round_total)

    return summary
```
